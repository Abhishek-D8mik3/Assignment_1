```python
#3.  Describe hoo to access modify and delete elements in a list withth examples

# Accessing Elements in a List:
     #You can access elements in a list using indexing or slicing.
       #Indexing:
           #Use square brackets ([]) to access an element by its index.
             #Indices start from 0 (for the first element) and go up to n-1 (for the last element), where n is the length of the list.
my_list = [10, 20, 30, 40, 50]

# Accessing the first element
print(my_list[0])  # Output: 10

# Accessing the last element using negative indexing
print(my_list[-1])  # Output: 50



#Slicing:
       #Use a colon (:) to extract a portion of the list.
       #my_list[start:end]: This returns elements from index start to end-1
my_list = [10, 20, 30, 40, 50]

# Accessing elements from index 1 to 3
print(my_list[1:4])  # Output: [20, 30, 40]

#Modifying Elements in a List:
      #You can modify individual elements or a slice of elements by assigning new values.

      #Modifying a Single Element:
      #Simply assign a new value to the list at the desired index.
my_list = [10, 20, 30, 40, 50]

# Changing the second element (index 1)
my_list[1] = 25
print(my_list)  # Output: [10, 25, 30, 40, 50]


#Modifying a Slice of Elements:
#You can modify multiple elements at once by assigning a new list to a slice.

my_list = [10, 20, 30, 40, 50]

# Replacing elements from index 1 to 3
my_list[1:4] = [21, 31, 41]
print(my_list)  # Output: [10, 21, 31, 41, 50]



# Deleting Elements in a List:
       #You can delete elements from a list using the del statement, remove(), or pop().

        #Using del:
         #The del statement removes an element or a slice of elements by index.

 y_list = [10, 20, 30, 40, 50]

# Deleting the third element (index 2)
del my_list[2]
print(my_list)  # Output: [10, 20, 40, 50]

# Deleting a slice of elements (from index 1 to 2)
del my_list[1:3]
print(my_list)  # Output: [10, 50]


#Using remove():
      #The remove() method deletes the first occurrence of a specific value.
         #If the value doesn't exist, it raises a ValueError.

 my_list = [10, 20, 30, 40, 50]

# Removing the element with value 30
my_list.remove(30)
print(my_list)  # Output: [10, 20, 40, 50]


#Using pop():
     #The pop() method removes and returns an element at a specific index.
      #If no index is specified, it removes and returns the last element.
  y_list = [10, 20, 30, 40, 50]

# Popping (removing) the last element
last_element = my_list.pop()
print(last_element)  # Output: 50
print(my_list)  # Output: [10, 20, 30, 40]

# Popping the second element (index 1)
second_element = my_list.pop(1)
print(second_element)  # Output: 20
print(my_list)  # Output: [10, 30, 40]
```

```python
#5  Describe the key features of sets and provide examples of their use
#Key Features of Sets in Python:
    #Unordered: Sets do not maintain the order of elements.
    #Unique Elements: Sets automatically remove duplicates. Each element is unique.
    #Mutable: You can add or remove elements from a set, but individual elements inside the set must be immutable (e.g., numbers, strings, tuples).
    #No Indexing: Since sets are unordered, elements cannot be accessed by index.
     #Set Operations: Sets support mathematical operations like union, intersection, difference, and symmetric difference.
# Creating a set
my_set = {1, 2, 3, 4, 4}  # Duplicate 4 will be removed
print(my_set)  # Output: {1, 2, 3, 4}

# Adding an element
my_set.add(5)
print(my_set)  # Output: {1, 2, 3, 4, 5}

# Set operations
set_a = {1, 2, 3}
set_b = {3, 4, 5}

# Union (elements in either set)
print(set_a | set_b)  # Output: {1, 2, 3, 4, 5}

# Intersection (common elements)
print(set_a & set_b)  # Output: {3}
```

```python
#7 Describe how to add modify and delete items in a dictionary with examples
#
 #Adding Items to a Dictionary:
   #You can add new key-value pairs by assigning a value to a new key.
  #Example:
my_dict = {'name': 'Alice', 'age': 25}
my_dict['city'] = 'New York'  # Adding a new key-value pair
print(my_dict)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}


#Modifying Items in a Dictionary:
   #Modify an existing value by re-assigning a value to an existing key.
  #Example:
my_dict = {'name': 'Alice', 'age': 25}
my_dict['age'] = 30  # Modifying the value of an existing key
print(my_dict)  # Output: {'name': 'Alice', 'age': 30}


#3. Deleting Items from a Dictionary:
#Use del to remove a specific key-value pair or the pop() method to remove and return a value.
#Example:


my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}

# Using del
del my_dict['city']
print(my_dict)  # Output: {'name': 'Alice', 'age': 25}

# Using pop
age = my_dict.pop('age')
print(my_dict)  # Output: {'name': 'Alice'}
print(age)  # Output: 25
#
```
